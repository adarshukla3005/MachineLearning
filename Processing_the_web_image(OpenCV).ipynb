**Libraries that can be used for image processing**

**Resizing the image using Pillow library**

```python
from PIL import Image
```

```python
img = Image.open('/content/Car.jpg')
img_resized = img.resize((200,200))
```

**Saving the image**

```python
img_resized.save('car_image_resized.jpg')
```

```python
# To read the image and convert it in numpy array
img_res = mpimg.imread('/content/car_image_resized.jpg')
```

```python
type(img_res)
```

```python
# displaying the image from numpy array

imp_plot = plt.imshow(img_res)
plt.show()
```

```python
print(img_res.shape)
```

**Now using OpenCV to read RGB image and convert into Grayscale **

```python
#importing IpenCV library
import cv2
```

```python
img = cv2.imread('/content/Car.jpg')
```

```python
type(img)
```

```python
img.shape
```

```python
grayscale_img = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
```

```python
type(grayscale_img)
```

```python
grayscale_img.shape
```
