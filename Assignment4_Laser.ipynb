```python
import numpy as np
import matplotlib.pyplot as plt

delta = np.linspace(0, 20, 1000)

R_values = [0.005,0.05, 0.2, 0.5, 0.8, 0.99]

plt.figure(figsize=(10, 6))
for R in R_values:
    T = 1 / (1 + (4 * R / (1 - R)**2) * np.sin(delta / 2)**2)
    plt.plot(delta, T, label=f'R={R}')

plt.xlabel('Phase difference δ (radians)')
plt.ylabel('Transmittance (T)')
plt.title('Transmission Characteristics of FPE')
plt.legend()
plt.grid(True)
plt.show()
```
